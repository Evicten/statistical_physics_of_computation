## Part I: 2.1 Gaussian Poincaré Inequality and Efron-Stein

### 1.

For the first equality we have:

$$Var[f(X)] = E[f(X)^2]-E[f(X)]^2$$

Now, since X' and X are identically distributed, we can write:

$$E[f(X)]^2 = E[f(X)]*E[f(X)] = E[f(X)]*E[f(X´)]$$

putting it together, we get:

$$Var[f(x)] = E[f(X)^2] - E[f(X)]*E[f(X')]$$

And since X and X' are independent:

$$Var[f(x)] = E[f(X)^2] - E[f(X)*f(X')] = E[f(X)^2 - f(X)*f(X')] = E[f(X)(f(X) - f(X'))]$$

Now, for the second equality, note that:

$$\sum_{i=1}^n{f(X_{(i-1)})-f(X_{(i)})} = f(X_{(0)}) - f(X_{(1)}) + f(X_{(1)}) - f(X_{(2)}) + ... - f(X_{(n-1)}) + f(X_{(n-1)})-f(X_{(n)}) = f(X) - f(X')$$

Therefore, 

$$E[f(X)(f(X) - f(X'))] = E[f(X)\sum_{i=1}^n{f(X_{(i-1)})-f(X_{(i)})}] = \sum_{i=1}^n{E[f(X)f(X_{(i-1)})-f(X_{(i)})]}$$

where the last step follows from the linearity of the expected value.

### 2.

Since X and X' are independent and identically distributed, we can make a change of variable $$ X_i <-> X'_i$$ in the expected value and get the same result. Using this, we get: $ X \to X^{(i)}$, $X_{(i-1)} \to X_{(i)}$ and $X_{(i)} \to X_{(i-1)} $

So:

$$E[f(X)f(X_{(i-1)})-f(X_{(i)})] = E[f(X^{(i)})f(X_{(i)})-f(X_{(i-1)})]$$

and summing them together:

$$E[f(X)f(X_{(i-1)})-f(X_{(i)})] = \frac{1}{2}E[f(X)f(X_{(i-1)})-f(X_{(i)})+f(X^{(i)})f(X_{(i)})-f(X_{(i-1)})] = \frac{1}{2}E[(f(X)-f(X^{(i)}))(f(X_{(i-1)})-f(X_{(i)}))]$$

### 3.

Cauchy-Schwartz states that:

$$|E[X*Y]|^2 \leq E[X^2]*E[Y^2]$$

Using Cauchy-Schwartz on the last expression on the previous exercise:

$$|E[(f(X)-f(X^{(i)}))*(f(X_{(i-1)})-f(X_{(i)}))]|^2 \leq E[(f(X)-f(X^{(i)}))^2]*E[(f(X_{(i-1)})-f(X_{(i)}))^2]$$

$$|E[(f(X)-f(X^{(i)}))*(f(X_{(i-1)})-f(X_{(i)}))]| \leq \sqrt{E[(f(X)-f(X^{(i)}))^2]*E[(f(X_{(i-1)})-f(X_{(i)}))^2]}$$

Now, if we take $E[(f(X_{(i-1)})-f(X_{(i)}))^2]$ and for any $i$ switch $X'_1, X'_2, ... , X'_{i-1}$ with $X_1, X_2, ... , X_{i-1}$ we should, again get the same expected value. Using this change of variable:

$$ E[(f(X_{(i-1)})-f(X_{(i)}))^2] = E[(f(X)-f(X^{(i)}))^2] $$

and putting it all together:

$$|E[(f(X)-f(X^{(i)}))*(f(X_{(i-1)})-f(X_{(i)}))]| \leq |E[(f(X)-f(X^{(i)}))^2]|$$

Therefore using the result from the previous exercise:

$$|E[f(X)f(X_{(i-1)})-f(X_{(i)})]| \leq \frac{1}{2}E[(f(X)-f(X^{(i)}))^2] $$

And finally we use the identity from the first exercise to prove Efron-Stein:

$$ Var[f(X)] = \sum_{i=1}^n{E[f(X)f(X_{(i-1)})-f(X_{(i)})]} \leq \frac{1}{2}\sum_{i=1}^n{E[(f(X)-f(X^{(i)}))^2]} $$

where we used the linearity of the expected value once again to take the sum out




### 4.

By Efron-Strein:

$$Var[f(\frac{S_n}{\sqrt{n}})] \leq \frac{1}{2}\sum_{i=1}^n{E[(f(\frac{S_n}{\sqrt{n}})-f(\frac{S_n^{(i)}}{\sqrt{n}}))^2]} $$

Let us examine the summand:

$$E[(f(\frac{S_n}{\sqrt{n}})-f(\frac{S_n^{(i)}}{\sqrt{n}}))^2]$$

For any $i$ we can split the random variable $\frac{S_n}{\sqrt{n}} = \frac{S_{n-1}}{\sqrt{n}} + \frac{X_i}{\sqrt{n}}$ and, likewise $\frac{S_n^{(i)}}{\sqrt{n}} = \frac{S_{n-1}}{\sqrt{n}} + \frac{X'_i}{\sqrt{n}}$ where $S_{n-1}$ is the same for both because all the random variables except $X_i$ are the same for both sums.

Then, we take the expected value only with respect to $X_i$ and $X'_i$:

$$ E_{X^{n-1}, X_i, X'_i}[(f(\frac{S_n}{\sqrt{n}})-f(\frac{S_n^{(i)}}{\sqrt{n}}))^2] = E_{X^{n-1}, X_i, X'_i}[(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{X_i}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} + \frac{X'_i}{\sqrt{n}}))^2]$$

$$=E_{X^{n-1}}[\frac{1}{4}(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}}))^2 + \frac{1}{4}(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} + \frac{-1}{\sqrt{n}}))^2 + \frac{1}{4}(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{-1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}}))^2 + \frac{1}{4}(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{-1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} + \frac{-1}{\sqrt{n}}))^2] $$

Two of the terms are 0, and the other two are the same, simplifying:

$$=E_{X^{n-1}}[\frac{1}{2}(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} - \frac{1}{\sqrt{n}}))^2] $$

And now we have removed the dependence on $i$ and we can put everything together neatly applying the sum over $i$ in the original expression:

$$Var[f(\frac{S_n}{\sqrt{n}})] \leq \frac{n}{4}E_{X^{n-1}}[(f(\frac{S_{n-1}}{\sqrt{n}} + \frac{1}{\sqrt{n}})-f(\frac{S_{n-1}}{\sqrt{n}} - \frac{1}{\sqrt{n}}))^2]$$

When it comes to the right hand side of our previous expression, we can send $\frac{n}{4}$ inside:

## Part II: 2.3 Mean-Field Algorithm and State Evolution of the RFIM

### 1. Cavity Method

The Hamiltonian of the RFIM is:

$$ \mathcal{H}_N(\vec{s}, \vec{h}) = -\frac{N}{2} \sum_{i}{(\frac{S_i}{N})^2}  - \sum_i h_i S_i $$

We can apply the cavity method in the exact same way as in the lecture notes, but for this Hamiltonian. We add one variable to a new system and compute the expect value of its spin in the new system as a function of the expected values in the old system. We start by writing:

$$-\beta'\mathcal{H}_{N+1} = \beta'\frac{1}{2}(N+1)(\frac{S_0+\sum_i{S_i}}{N+1})^2+\beta'(h_0S_0+\sum_i{h_iS_i})$$

Which expands into:

$$ -\beta'\mathcal{H}_{N+1} = o(1)+\beta'\frac{N}{N+1}S_0\frac{\sum_i{S_i}}{N}+\beta'\frac{N^2}{2(N+1)}(\frac{\sum_i{S_i}}{N})^2+\beta'h_0S_0+\beta'\sum_i{h_iS_i}$$

We can define $\beta' = \frac{N+1}{N}\beta$ and take $N \to \infty$:

$$ -\beta\mathcal{H}_{N+1} = \beta S_0\frac{\sum_i{S_i}}{N}+\beta \frac{N}{2}(\frac{\sum_i{S_i}}{N})^2+\beta h_0S_0+\beta\sum_i{h_iS_i}$$

Which is

$$ -\beta\mathcal{H}_{N+1} = -\beta\mathcal{H}_N + \beta S_0\frac{\sum_i{S_i}}{N} + \beta h_0S_0$$

The cavity method adds a linking term and a site term to the original hamiltonian. Hence, we get:

$$\langle S_0 \rangle_{N+1, \beta} = \frac{\sum_{S_0, \vec{S}}{S_0\exp{-\beta\mathcal{H}_{N+1}}}}{\sum_{S_0, \vec{S}}{\exp{-\beta\mathcal{H}_{N+1}}}} = \frac{\langle \sinh{(\beta(\bar{S}+h_0))} \rangle_{N, \beta}}{\langle \cosh{(\beta(\bar{S}+h_0))} \rangle_{N, \beta}}$$

Finally, we assume $\bar{S}$ concentrates on the same value $m^*$ for both the $N$ and $N+1$ systems in this large N limit:

$$\langle S_0 \rangle_{N+1, \beta} = \frac{ \sinh{(\beta(m^*+h_0))}}{\cosh{(\beta(m^*+h_0))}} = \tanh(\beta(m^*+h_0)) $$

In the zero temperature limit $\beta \to \infty$, this becomes:

$$\langle S_0 \rangle_{N+1, \beta} = sign(m^*+h_0)$$

Which is either -1 or 1, depending on the specific value of $h_0$. Therefore, noting that this argument can be made for any spin $S_i$ on the system, it is reasonable to implement the following iterative algorithm:

$$S_i^{t+1} = sign(h_i + \frac{\sum_i{S_i^t}}{N})$$



### 2. Implementing the algorithm and comparing with replica prediction

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

### Replica Prediction

First, I define the functions to compute the fixed-point iteration for finding the magnetization $m^*$, using the replica prediction:

$$m = \int{dh\frac{\exp{-\frac{h^2}{2\Delta}}}{\sqrt{2\pi\Delta}}}\tanh(\beta(h+m))$$

Which, in the limit $\beta \to \infty$ becomes:

$$m = \int{dh\frac{\exp{-\frac{h^2}{2\Delta}}}{\sqrt{2\pi\Delta}}}sign(h+m)$$

```python
# define the integrand function
def self_consistent_integrand(h, m, delta):
    """
    arguments:
        - h : external field
        - m : magnetization
        - delta : variance of the distribution of the random fields h
    returns:
        - the result of the integrand for the specific values of h and m
    """
    return np.exp(-h**2/(2*delta))/np.sqrt(2*np.pi*delta)*np.sign(m+h)

# compute the integral
def fixed_point_function(m, delta):
    """
    arguments:
        - m : magnetization
        - delta : variance of the distribution of the random fields h
    returns:
        - the result of the integral
    """
    integral, err = quad(self_consistent_integrand, -20, 20, args=(m, delta)) ## one could use np.inf for the limits, but after some tests it was found that using a limit such as 20 works fine
    return integral

# compute the fixed point iteration of the self consistent equation for m*
def fixed_point_iter(delta, m_init=0.4, eta=0.8, tol=1e-6, maxIter = 1e5):
    """
    arguments:
        - delta : variance of the distribution of the random fields h
        - m_init : initial value of the magnetization in the iteration
        - eta : damping parameter
        - tol : tolerance for the convergence of the fixed point iteration
        - maxIter : maximum number of iterations
    returns:
        - m : the fixed point of the self consistent equation
    """
    m = m_init
    m_prev = m_init
    for _ in np.arange(maxIter):
        m_prev = m
        m = fixed_point_function(m, delta)*eta + (1-eta)*m_prev
        if np.abs(m-m_prev) < tol:
            break
    return m
```

Then, I define the functions needed to compute the replica theoretical prediction for the minimum energy, which is (from the lecture notes):

$$E[e_{min}] = \frac{(m^*)^2}{2}-E_h[(h+m^*)sign(h + m^*)]$$

```python
# define the integrand function
def replica_integrand(h, delta, mstar):
    """
    arguments:
        - h : external field
        - delta : variance of the distribution of the random fields h
        - mstar: the fixed point magnetization
    returns:
        - the result of the integrand
    """
    return np.exp(-h**2/(2*delta))/np.sqrt(2*np.pi*delta)*(h+mstar)*np.sign(h+mstar)

#compute the replica minimum energy
def replica_min_energy(delta):
    """
    arguments:
        - delta : variance of the distribution of the random fields h
    returns:
        - the replica minimum energy
    """
    mstar = fixed_point_iter(delta)
    h_expected_value, err = quad(replica_integrand, -20, 20, args=(delta, mstar))
    return mstar**2/2 - h_expected_value
```

Before moving on, we make some sanity checks that both these replica predictions computations are working as expected:

```python
delta_space = np.linspace(0.1, 1, 100)
magnetizations_plus = []
magnetizations_minus = []
for delta in delta_space:
    magnetizations_plus.append(fixed_point_iter(delta, m_init=0.4))
    magnetizations_minus.append(fixed_point_iter(delta, m_init=-0.4))

plt.plot(delta_space, magnetizations_plus, label='m* for m_init=0.4')
plt.plot(delta_space, magnetizations_minus, label='m* for m_init=-0.4')
plt.xlabel('$\Delta$')
plt.ylabel('m*')
plt.title('Fixed point iteration results as function of $\Delta$')
plt.legend()
plt.show()
```

This is expected. For $\Delta$ near 0 the effect of the random fields is negligible and we recover the zero temperature behaviour of Curie-Weiss model with h = 0. As $\Delta$ increases, it becomes increasingly favorable for the spins to align themselves with the specific values of $h_i$ than to each other, and the magnetization becomes closer to 0. 

Next we plot the replica minimum energy prediction:

```python
delta_space = np.linspace(0.01, 1, 200)
energies = []
for delta in delta_space:
    energies.append(replica_min_energy(delta))

plt.plot(np.sqrt(delta_space), energies)
plt.xlabel(r'$\sqrt{\Delta}$')
plt.ylabel(r'$E_{min}$')   
plt.title('Replica minimum energy as function of $\sqrt{\Delta}$') 
plt.legend()
plt.show()
```

And I get the same behaviour as expected. Note that I used the standard deviation in the x-axis, instead of the variance

### Algorithm

Now, we finally code the algorithm.

```python
#initialize a random configuration of N spins
def initialize_spins(N):
    """
    arguments:
        - N : number of spins to initialize
    returns:
        - an array of N spins randomly chosen between 1 and -1
    """
    return np.random.choice([-1,1], N)

#compute the energy density of a configuration of spins
def energy_density(s, h):
    """
    arguments:
        - s : the configuration of spins - array of dimension (N,1) 
        - h : the external random field configuration - array of dimension (N,1)
    returns:
        - the energy density of the configuration with given h external random fields
    """
    return -1/2*(np.mean(s))**2-np.mean(np.multiply(h,s))
```

```python
# run the algorithm
def algo(h, maxIter = 100):
    """
    arguments:
        - h : the external random field configuration - array of dimension (N,1)
        - maxIter : maximum number of iterations 
    returns:
        - the final configuration of spins after maxIter iterations
    """
    N = len(h)
    s = initialize_spins(N)
    for _ in range(maxIter):
        s = np.sign(h+np.mean(s))

    return s
```

```python
N = 10000
delta_sqrt_space = np.linspace(0.1, 1, 200)

replica_energies = []
for delta_sqrt in delta_sqrt_space:
    replica_energies.append(replica_min_energy(delta_sqrt**2))

algo_energies = []
for delta_sqrt in delta_sqrt_space:
    h = np.random.normal(0, delta_sqrt, N)
    s  = algo(h, maxIter=100)    
    algo_energies.append(energy_density(s, h))

plt.plot(delta_sqrt_space, algo_energies, label='algorithm result, N = 10000')
plt.plot(delta_sqrt_space, replica_energies, label='replica prediction')
plt.xlabel(r'$\sqrt{\Delta}$')
plt.ylabel(r'$E_{min}$')
plt.title('Replica minimum energy and algorithm result as function of $\sqrt{\Delta}$')
plt.legend()
plt.show()
```

The algorithm behaves accordingly to the replica prediction, with some oscillations. We shall now study the algorithm results for different values of N:

```python
N_list = [500, 1000, 2000, 5000, 10000]
delta_sqrt_space = np.linspace(0.1, 1, 200)

replica_energies = []
for delta_sqrt in delta_sqrt_space:
    replica_energies.append(replica_min_energy(delta_sqrt**2))

algo_energies_all = []

for N in N_list:
    algo_energies = []
    for delta_sqrt in delta_sqrt_space:
        h = np.random.normal(0, delta_sqrt, N)
        s  = algo(h, maxIter=300)    
        algo_energies.append(energy_density(s, h))
    algo_energies_all.append(algo_energies)

for idx, N in enumerate(N_list):
    plt.plot(delta_sqrt_space, algo_energies_all[idx], label='algorithm result, N = {}'.format(N), alpha=0.1*(idx+4))

plt.plot(delta_sqrt_space, replica_energies, label='replica prediction')
plt.xlabel(r'$\sqrt{\Delta}$')
plt.ylabel(r'$E_{min}$')
plt.title('Replica minimum energy and algorithm result as function of $\sqrt{\Delta}$')
plt.legend()
plt.show()
```

In general, the larger the N, the less "spikes" the algorithm has outside its predicted behaviour

### 3. State Evolution Equation

We already have everything we need to compare the algorithm's behaviour with the replica fixed point equation for the magnetization. First, we compare the analytics and then we shall plot both their evolution in time.

The algorithm we're using is:

$$S_i^{t+1} = sign(h_i + \frac{\sum_i{S_i^t}}{N})$$

Averaging over all spins we get:

$$m^{t+1} = \sum_i{\frac{sign(h_i + m^t)}{N}}$$

In the large N limit, since we are sampling h_i from a distribution, this will correspond to the expected value in h:

$$m^{t+1} = E_h[sign(h+m^t)]$$

Which is exactly the replica self consistent equation for m. Therefore, for large N the algorithm is solving this self consistent equation by a fixed point iteration, which is equivalent to solving the replica symmetric free energy equation.

Finally, we plot both evolutions for $\Delta = 0.6$ as an example:

```python
## slightly changed fixed_point_iter to return all time evolution
def fixed_point_iter_time(delta, m_init=0.4, eta=0.8, tol=1e-6, maxIter = 1e5):
    """
    arguments:
        - delta : variance of the distribution of the random fields h
        - m_init : initial value of the magnetization in the iteration
        - eta : damping parameter
        - tol : tolerance for the convergence of the fixed point iteration
        - maxIter : maximum number of iterations
    returns:
        - m : the fixed point of the self consistent equation
    """
    m = m_init
    m_prev = m_init
    m_list = []
    for _ in np.arange(maxIter):
        m_prev = m
        m = fixed_point_function(m, delta)*eta + (1-eta)*m_prev
        m_list.append(m)
    return m_list

## algorithm but with recorded time evolution of m
def algo_time(s_init, h, maxIter = 100):
    """
    arguments:
        - s_init : the initial configuration of spins - array of dimension (N,1)
        - h : the external random field configuration - array of dimension (N,1)
        - maxIter : maximum number of iterations 
    returns:
        - s: the final configuration of spins after maxIter iterations
        - m_list: the time evolution of the magnetization

    """
    N = len(h)
    s = s_init
    m_list = []
    for _ in range(maxIter):
        s = np.sign(h+np.mean(s))
        m_list.append(np.mean(s))

    return s, m_list
```

```python
## plot the time evolution of m for replica fixed point iteration and algorithm, for some delta

N = 1000000
delta = 0.6
s = np.ones(N)
h = np.random.normal(0, np.sqrt(delta), N)

plt.plot(fixed_point_iter_time(delta, m_init=np.mean(s), maxIter=500), label='fixed point m evolution')
s, m_list = algo_time(s, h, maxIter=500)
plt.plot(m_list, label = 'algorithm m evolution')
plt.legend()
plt.show()
```

They both exhibit the same behaviour, as expected, albeit with a slight difference. I would assume that, with N larger, the difference becomes smaller and smaller.
